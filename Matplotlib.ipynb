```python
import matplotlib.pyplot as mp
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
mp.plot(x,y)


```

```python
fig = mp.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('TITLE')
```

```python
fig,axes = mp.subplots(nrows = 1, ncols= 2 )
mp.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x)
```
