```python
import tensorflow as tf
import sklearn
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\jayad\Desktop\DATA\Titanic\train.csv')
df_test = pd.read_csv(r'C:\Users\jayad\Desktop\DATA\Titanic\test.csv')
```

```python
df = df.fillna(0)
```

```python
df = df.drop('Ticket', axis=1).drop('Cabin', axis=1).drop('Name', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
df['Embarked'] = enc.fit_transform(df['Embarked'].astype(str))
```

```python
df['Sex'] = enc.fit_transform(df['Sex'].astype(str))
```

```python
data_in = df.drop("Survived", axis=1)
data_in = data_in.values
```

```python
labels = df['Survived']
```

```python
data_in 
```

```python
labels = labels.astype("float64")
```

```python
labels.size
```

```python
data_in.size
```

```python
from tensorflow.keras import layers
```

```python
model = tf.keras.Sequential()
```

```python
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(80, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(0.001),loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(data_in, labels, batch_size=50, epochs=10)
```
